```python
#Task 1

class Student:
    ID = 0
    def __init__(self,name,department,age,cgpa):
        self.name = name
        self.department = department
        self.age = age
        self.cgpa = cgpa
        Student.ID += 1
        self.ID = Student.ID
    def get_details(self):
        print(f"ID: {self.ID}")
        print(f"Name: {self.name} "
              f"\nDepartment: {self.department} "
              f"\nAge: {self.age} "
              f"\nCGPA: {self.cgpa}")
    @staticmethod
    def from_String(string):
        b = list(str(x) for x in string.split("-"))
        return Student(b[0], b[1], b[2], b[3])





s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

print('\n')
print('Instance variables are used for storing data in a particular instance and Class variables are variables that is shared between all instances of a class')

print('\n')

print("Instance methods need a class instance that can access the instance through self. Class methods don't need a class instance")
```

```python
#Task 2

class Assassin:
    count = 0
    def __init__(self,name,sucess_rate):
        self.name = name
        self.success_rate = sucess_rate
        Assassin.count += 1
    def printDetails(self):
        print(f"Name: {self.name}"
              f"\nSuccess Rate: {self.success_rate}%"
              f"\nTotal number of Assassin: {Assassin.count}")
    @classmethod
    def failureRate(cls,name,rate):
        rate = 100 - rate
        return cls(name, rate)

    @classmethod
    def failurePercentage(cls,name, rate):
        rate = 100-rate
        return cls(name,rate)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task 5

class Employee:
    def __init__(self,name,year):
        self.name = name
        self.workingPeriod = year
    @classmethod
    def employeeByJoiningYear(cls,x,y):
        return Employee(x,2021-y)

    @staticmethod
    def experienceCheck(year,gender):
        if gender=="male":
            if year<3:
                return 'She is not experienced'
            else:
                return 'She is experienced'
        elif gender=='female':
            if year<3:
                return 'He is not experienced'
            else:
                return 'He is experienced'

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 6

class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount += self.count
    @staticmethod
    def advantage():
        print("Laptops are portable")

    @staticmethod
    def resetCount():
        Laptop.laptopCount = 0


lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#Task 7

class Cat:
    Number_of_cats = 0
    def __init__(self,color,do):
        self.color= color
        self.do = do
        Cat.Number_of_cats += 1

    @classmethod
    def no_parameter(cls,color = 'White',do = 'sitting'):
        return cls(color,do)

    @classmethod
    def first_parameter(cls,color,do='sitting'):
        return cls(color,do)
    @classmethod
    def second_parameter(cls,do,color='Grey'):
        return cls(color,do)

    def changeColor(self,color):
        self.color = color

    def printCat(self):
        print(f'{self.color} cat is {self.do}')





print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 8

class Cylinder:
    radius = 5
    height = 18

    def __init__(self,radius,height):
        self.radius = radius
        self.height = height
        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}')
        print(f'Updated: radius={self.radius} and height={self.height}')
        Cylinder.radius = self.radius
        Cylinder.height = self.height

    @staticmethod
    def area(x,y):
        total = 2*3.14159265359*(x*x) + 2*3.14159265359*x*y
        print(f"Area: {total}")
    @staticmethod
    def volume(x,y):
        total = 3.14159265359*(x**2)*y
        print(f"Volume: {total}")

    @classmethod
    def swap(cls,radius,height):
        return cls(height,radius)

    @classmethod
    def changeFormat(cls,a):
        b = list(str(x) for x in a.split("-"))
        return cls(int(b[0]),int(b[1]))
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)

```

```python
#Task 10

class SultansDine:
    Branches = 0
    Total = 0
    a = []
    b = []
    c = []
    def __init__(self,Branch_name):
        self.Branch_name = Branch_name
        self.Branch_sell = 0
        self.total = 0
        SultansDine.Branches += 1

    def sellQuantity(self,x):
        if x <10:
            self.Branch_sell = x*300
        elif x <20:
            self.Branch_sell = x*350
        else:
            self.Branch_sell = x*400

        SultansDine.Total += self.Branch_sell
        SultansDine.b.append(self.Branch_sell)
        SultansDine.a.append(self.Branch_name)
        SultansDine.c.append(x)
    def branchInformation(self):
        print(f"Branch Name: {self.Branch_name}")
        print(f"Branch Sell: {self.Branch_sell}")

    @staticmethod
    def details():
        print(f"Total Number of branch(s): {SultansDine.Branches}")
        print(f"Total Sell: {SultansDine.Total}")
        if SultansDine.Branches == 0:
            pass
        else:
            for i in range(SultansDine.Branches):
                print(f'Branch Name: {SultansDine.a[i]}, Branch Sell: {SultansDine.b[i]}')
                print(f"Branch consists of total sells's: {round((SultansDine.b[i]/SultansDine.Total)*100,2)}%")




SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
